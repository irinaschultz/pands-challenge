```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(8)
```

```python
#Calculate the tital number of schools
total_schools=school_data["school_name"].count()
total_schools
```

```python
#Calculate the total number of student
total_students=student_data["student_name"].count()
total_students
```

```python
#calculate the total budget
total_budget=school_data["budget"].sum()
total_budget
```

```python
#calculate the averadge maths score
average_math=student_data[("maths_score")].mean()
average_math
```

```python
#calculate the average reading score
average_reading=student_data[("reading_score")].mean()
average_reading
```

```python
#calculate the percentage of student with a passing maths score (50 or greater)
maths_pass=(school_data_complete[school_data_complete["maths_score"]>=50].maths_score.count())/total_students
maths_pass_per=maths_pass*100
maths_pass_per
```

```python
#calculate the percentage of student with a passing score (50 or greater)
reading_pass=(school_data_complete[school_data_complete["reading_score"]>=50].reading_score.count())/total_students
reading_pass_per=reading_pass*100
reading_pass_per
```

```python
#calculate the students who passed maths and reading
pass_maths_reading=school_data_complete[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)]
#counting the students who passed maths and reading
pass_maths_reading_count = pass_maths_reading["student_name"].count()
##calculate the percentage of student who passed maths and reading(%overall passing)
overall_pass=pass_maths_reading_count/total_students*100
overall_pass
```

```python
#list of values with keys to create a new DataFrame.
district_summary_df = pd.DataFrame(
          [{"Total Schools": total_schools,
          "Total Students": total_students,
          "Total Budget": total_budget,
          "Average Math Score": average_math,
          "Average Reading Score": average_reading,
          "% Passing Math": maths_pass_per,
         "% Passing Reading": reading_pass_per,
        "% Overall Passing Rate": overall_pass}])
district_summary_df
```

```python
#display the cleaner formating /# Reformat Total Students and Total Budget

district_summary_df['Total Students'] = district_summary_df.apply(lambda x: "{:,.0f}".format(x['Total Students']), axis=1)

district_summary_df['Total Budget'] = district_summary_df.apply(lambda x: "${:,.2f}".format(x['Total Budget']), axis=1)

district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)

district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)

district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)

district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)

district_summary_df["% Overall Passing Rate"] = district_summary_df["% Overall Passing Rate"].map("{:.0f}".format)


district_summary_df
```

```python
#school mame 
school_summary = school_data.groupby(['school_name'])
school_names = school_data.school_name.sort_values().unique() 
```

```python
#school types
school_types = school_data.set_index(["school_name"])["type"]
df = pd.DataFrame(school_types)
```

```python
#total students per school 
students_counts = ["size"]

students_counts = school_data.set_index(["school_name"])["size"]

# Calculate the total student count.
students_counts = school_data_complete["school_name"].value_counts()

```

```python
# Calculate the total school budget.
per_school_budget = school_data.set_index(["school_name"])["budget"]

# budget per student
per_student_budget = per_school_budget / students_counts

```

```python
#school_avg maths+reading scores

per_school_math = school_data_complete.groupby(["school_name"]).mean()["maths_score"]
per_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
```

```python
#the passing scores by filtered DataFrame.
per_school_passing_maths = school_data_complete[(school_data_complete["maths_score"] >= 50)]

per_school_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 50)]

#number of students passing math &  reading by school.
per_school_passing_math = per_school_passing_maths.groupby(["school_name"]).count()["student_name"]

per_school_passing_reading = per_school_passing_reading.groupby(["school_name"]).count()["student_name"]

# % passing maths & reading  per school.
per_school_passing_math = per_school_passing_math / students_counts * 100

per_school_passing_reading = per_school_passing_reading / students_counts * 100

# number of students who passed maths and reading.
per_passing_math_reading = school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)]

# number of students who passed maths and reading per school.
per_passing_math_reading = per_passing_math_reading.groupby(["school_name"]).count()["student_name"]

# % the overall passing 
per_overall_passing_percentage = per_passing_math_reading / students_counts* 100
```

```python
#new DataFrame.

new_school_summary_df = pd.DataFrame({
             "School Type": school_types,
             "Total Students": students_counts,
             "Total School Budget": per_school_budget,
             "Per Student Budget": per_student_budget,
             "Average Math Score": per_school_math,
             "Average Reading Score": per_school_reading,
             "% Passing Math": per_school_passing_math,
             "% Passing Reading": per_school_passing_reading,
             "% Overall Passing Rate": per_overall_passing_percentage})

# Format school budget column and per student budget column.
new_school_summary_df["Total School Budget"] = new_school_summary_df["Total School Budget"].map("${:,.2f}".format)

new_school_summary_df["Per Student Budget"] = new_school_summary_df["Per Student Budget"].map("${:,.2f}".format)


# Display the df
new_school_summary_df

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
new_school_summary_df.head()
```

```python
 #create the bins
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

new_school_summary_df["Spending Summary (Per student)"] = pd.cut(per_student_budget, spending_bins, labels=group_names)
scores_by_sch=new_school_summary_df.drop(columns=["Total Students", "Total School Budget","School Type",])
scores_by_sch
```

```python
# Calculate averages columns. 
spending_math_scores = scores_by_sch.groupby(["Spending Summary (Per student)"]).mean()["Average Math Score"]

spending_reading_scores = scores_by_sch.groupby(["Spending Summary (Per student)"]).mean()["Average Reading Score"]

spending_passing_math = scores_by_sch.groupby(["Spending Summary (Per student)"]).mean()["% Passing Math"]

spending_passing_reading = scores_by_sch.groupby(["Spending Summary (Per student)"]).mean()["% Passing Reading"]

overall_passing_spending = scores_by_sch.groupby(["Spending Summary (Per student)"]).mean()["% Overall Passing Rate"]

#Create the DataFrame
spending_summary_df = pd.DataFrame({
          "Average Math Score" : spending_math_scores,
          "Average Reading Score": spending_reading_scores,
          "% Passing Math": spending_passing_math,
          "% Passing Reading": spending_passing_reading,
          "% Overall Passing": overall_passing_spending})
spending_summary_df

```
